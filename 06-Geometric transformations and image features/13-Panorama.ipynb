# Steps for Creating Panoramas

1. Find keypoints in all images
2. Find pairwise correspondences
2. Estimate pairwise Homographies
3. Refine Homographies
3. Stitch with Blending

## 1- Importing Libraries

```python
import cv2
import glob
import math
import matplotlib.pyplot as plt
```

## 2- Loading Images

```python
imagefiles = glob.glob("images/boat*.jpg")
imagefiles.sort()

images = []
for filename in imagefiles:
    img = cv2.imread(filename)
    images.append(img)

num_images = len(images)
num_images
```

## 3- Plotting Images

```python
plt.figure(figsize=[30,10])

num_cols = 3
num_rows = math.ceil(num_images / num_cols)

for i in range(0, num_images):
    plt.subplot(num_rows, num_cols, i+1) 
    plt.axis('off')
    plt.imshow(images[i][...,::-1])
```

## 4- Stitching

```python
stitcher = cv2.Stitcher_create()

status, result = stitcher.stitch(images)

if status == 0:
    plt.figure(figsize=[30,10]) 
    plt.imshow(result[...,::-1])
```
