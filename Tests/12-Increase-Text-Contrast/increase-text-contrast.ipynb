## 1- Importing Libaries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Global Histogram

```python
image = cv2.imread('hidden.jpg', 0)
result = cv2.equalizeHist(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("Global Histogram Equalization")
```
