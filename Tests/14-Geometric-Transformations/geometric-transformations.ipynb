## 1- Importing Libaries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loading Image

```python
image = cv2.imread('todd-trapani.jpg')
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
height, width, _ = img.shape
```

## 3- Combination of Translations

```python
angle = 30 
center = (width // 2, height // 2)


M_rotate = cv2.getRotationMatrix2D(center, angle, 1.0)

M_translate = np.float32([[1, 0, 10], [0, 1, 0]])

M_combined = np.dot(np.vstack([M_rotate, [0, 0, 1]]), np.vstack([M_translate, [0, 0, 1]]))

M_combined = M_combined[:2]


rotated_translated_image = cv2.warpAffine(img, M_combined, (width, height))
img_result = cv2.cvtColor(rotated_translated_image, cv2.COLOR_BGR2RGB)


plt.figure(figsize=[10,8])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(img_result[...,::-1]);plt.title("Result")
```
