## 1- Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loading Image

```python
img = cv.imread("Albert-Einstein.jpg")

plt.imshow(img)
plt.title("Original Image")
plt.show()
```

## 3- Cropping Face

```python
height, width = img.shape[:2]

face = img[250:850, 400:850]

plt.imshow(face)
plt.title("Face Image")
plt.show()
```

## 4- Blurring Cropped

```python
kernel = np.ones((30, 30), np.float32) / 900

blurred = cv.filter2D(face, -1, kernel)

plt.figure(figsize=[12,8])
plt.subplot(121);plt.imshow(face[...,::-1]);plt.xticks([]);plt.yticks([]);plt.title("Original Face")
plt.subplot(122);plt.imshow(blurred[...,::-1]);plt.xticks([]);plt.yticks([]);plt.title("Blurred Face")
```

## 5- Finall Image

```python
blurred_face_image = img.copy()
blurred_face_image[250:850, 400:850] = blurred

plt.imshow(blurred_face_image);plt.xticks([]);plt.yticks([]);plt.title("Face Blurred")
plt.show()
```
