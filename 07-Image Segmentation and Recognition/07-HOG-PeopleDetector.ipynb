## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Example-1

### 1- Creating a HOG-Descriptor Object

```python
hog = cv2.HOGDescriptor()

# Initialize the People Detector
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

### 2- Loading Image

```python
image = cv2.imread('images/pedestrian.jpg')
(bounding_boxes, weights) = hog.detectMultiScale(image,
                                                 winStride=(4,4),
                                                 scale=1.05)

# Draw bounding boxes on the image
for (x, y, w, h) in bounding_boxes:
    cv2.rectangle(image,
    (x,y),
    (x + w, y + h),
    (0, 255, 255),
    4)
    
plt.imshow(image[...,::-1])
```

## Example-2

### 1- Creating a HOG-Descriptor Object

```python
hog = cv2.HOGDescriptor()

# Initialize the People Detector
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

### 2- Loading Image

```python
image = cv2.imread('images/pedestrians_2.jpg')
(bounding_boxes, weights) = hog.detectMultiScale(image)

# Draw bounding boxes on the image
for (x, y, w, h) in bounding_boxes: 
    cv2.rectangle(image, 
                  (x, y),  
                  (x + w, y + h),  
                  (0, 255, 255), 
                   4)

plt.imshow(image[...,::-1])
```

- We usually set the optional scale parameter to a value between 1.01 and 1.5, which greatly affects the speed of the algorithm!

```python
%timeit hog.detectMultiScale(image, scale=1.01)
```

```python
%timeit hog.detectMultiScale(image, scale=1.5)
```
