## 1- Importing Libraries

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

## 2- Equalization

```cv2.equalizeHist(image)```

````python
image = cv2.imread('images/xray.jpg', 0)
result = cv2.equalizeHist(image)

plt.figure(figsize=[8,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original X-ray image")
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("Equalized Histogram")
````

### - Histogram

````python
image = cv2.imread('images/xray.jpg', 0)
result = cv2.equalizeHist(image)

plt.figure(figsize=[12,7])
plt.subplot(221);plt.imshow(image, cmap='gray');plt.xticks([]);plt.yticks([]);plt.title("Original X-ray image")
plt.subplot(222);plt.hist(image.ravel(), 256)

plt.subplot(223);plt.imshow(result, cmap='gray');plt.xticks([]);plt.yticks([]);plt.title("Equalized Histogram")
plt.subplot(224);plt.hist(result.ravel(), 256)

plt.show()
````

## 5- Colored Image

````python
image = cv2.imread('images/mid.png')

# Histogram Equalization
channels = cv2.split(image)
eq_channels = []
for ch in channels:
    eq_channels.append(cv2.equalizeHist(ch))
eq_image = cv2.merge(eq_channels)


#plot images
plt.figure(figsize=[12,7])
plt.subplot(221);plt.imshow(image[...,::-1]);plt.xticks([]);plt.yticks([]);plt.title("Original")
plt.subplot(223);plt.imshow(eq_image[...,::-1]);plt.xticks([]);plt.yticks([]);plt.title("Equalized")


color = ('b', 'g', 'r')
#Original image histogram
plt.subplot(222)
for i, col in enumerate(color):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
    
#Equualized image histogram
plt.subplot(224)
for i, col in enumerate(color):
    histogram = cv2.calcHist([eq_image], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
````
